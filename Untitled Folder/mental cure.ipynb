事情的起因是5月12日那一天，我又一次找了周宇琪，想和她聊聊天，然后和她很正式地道个歉，说我们不可能再在一起了。因为我知道即使我们在一起了，我也会再一次因为各种各样的事情，尤其是不在乎她感受、只考虑我自己感受的事情伤害到她。我已经非常笃定地告诉自己，不行。但是，发完道歉的短信之后，看到她回我说，她以为我是去道歉的，没想到我说了那些东西时，我又开始幻想了。当时我在想我们是不是可能还能在一起，是不是还能和好如初。然后我就告诉了七姐我还很喜欢她，但是也没和她说重新在一起，也没说我们不能在一起，因为我当时主要的想法还是道歉（现在想起来就真的很有病，没有把话说清楚）。然后就和她发了谴责我自己、和道歉的话，她发了一个暂时不会原谅我的，至此一切倒还算好，如果她不理我一切就都算好。

问题就是第二天，周宇琪加了我，问我问题啥的。我当时还尚有一点理智在，就想和她打个电话说一下我现在的想法。但听到她的声音，我又开始在想，我是不是可以和她在一起。虽然说了我们删微信吧，但是中间还是存在我们可以在一起的想法，于是话也没说清楚，就支支吾吾的。之后，我还问你你为什么不骂我，她说：“我为什么要骂你？”说罢，我心里就在想，你是不是讨厌我了，对我做错事情也不管。然后又说：“你和她在一起不就是为了馋她身子？我还不知道你么？”我心里又开始想，虽然其中一个原因是这个，但我心里的感受你就完全不顾了么？你又懂我什么？我很希望有个人能陪在我身边，当我的救命稻草不可以么？然后她就叫我删朋友圈，不理晓钰什么的。我这时候脑袋已经要爆炸了，没法思考任何东西。

当时我又想起了以前，周宇琪以一种很嘲讽的语气和我说话，说得她非常懂我，知道我心里在想些什么一样，但她却不明白怎么化解我心中的想法。我很明白她说的东西是对的，但是我听起来就是很不舒服，就是想让我不断地想去逃避她。我们在一起后期，都只是再聊一些有的没的，甚至就按周宇琪说的，我好像就是对她除了做爱就没有别的想法了（回想起来确实，那个时候已经除了性，我什么也不想去想了，因为性对我来说是简单快乐的，而别的都太复杂了，我害怕被伤害）。周宇琪说的东西很多都是对的，但是我没法理解。我很多时候听完她说的话之后，脑袋只剩下了她骂我的声音，还有无尽的，我自己对自己说的话：不可以生气，她是不是在伤害我，我还做错了什么，我为什么要这样。

感觉写得好乱，分几个小点写我的偏执+精神分裂的结果吧

* ### 感觉大家都在伤害我 ###

首先，这个大家不包括我不认为亲近的人。越亲近的人，我越容易有这种想法。啊，写到这里我想起来一件事情，超级无敌重要的，就是我小时候睡觉前总是会胡思乱想几十分钟，比如我化身为钢铁侠、超人什么的。但还有一个很重要的，姑且称之为晚日梦吧，就是我不断地被父母、姐姐、家人嫌弃，并且被打、被骂，被从楼上扔下去，全身脏兮兮的。明明他们没有怎么说我，怎么骂我，但我心中止不住会有这么想。跟周宇琪在一起的前面估计是没有这么想，但越到了后期，心里这种被伤害、被抛弃的感觉越来越强烈，以至于碰到一点小事情，我都想要离开周宇琪，害怕和她相处。

“你没有必要为未来可能发生的事情感到自责，因为不一定真的发生，而且你现在也在努力不让这些事情发生。其实我不会特别在意。你凶我当时可能会比较难过，但是很快能好。像是今天你生气了，后来你抱了抱我，我也就好了。”

在这之前确实我已经和晓钰经常闹了，原因也就是我心里一直想念着周宇琪，没有去遗忘她。晓钰叫我删照片、删空间、丢小星星，每一件事情做完之后我都跟疯了一样。特别是把小星星丢了的事情，我大脑简直是要爆炸了。
